```python
# 4.Using the methods described in this module, preprocess all numerical and categorical variables, as needed
#Create the OneHotEncoder instance
Merged_encoded = pd.get_dummies(merged_3Years_df, columns=["state", "floodzone", "occupancytype", "lossmonth"])
# Drop the null columns where all values are null
Merged_encoded = Merged_encoded.dropna(axis='columns', how='all')
Merged_encoded
```

```python
#Split our preprocessed data into our features and target arrays
y =  Merged_encoded['amountpaidonbuildingclaim'].values
X =  Merged_encoded.drop(['amountpaidonbuildingclaim'],1).values
#X_new= SelectKBest(chi2, k=20).fit_transform(X, y)
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)

# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

```

```python
X_train.shape
```

```python
#5.
# Define the model - deep neural net
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 4
#hidden_nodes_layer2 = 4

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))

# Second hidden layer
#nn.add(
#    tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(
    tf.keras.layers.Dense(units=1))#, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
#6.Compile, train, and evaluate your binary classification model.
# Compile the model
#nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
#nn.compile(loss="mse", optimizer="adam", metrics=["r2"])
nn.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
fit_model2 = nn.fit(X_train_scaled, y_train, epochs=5)
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(number_input_features, 
                input_dim=number_input_features, 
                kernel_initializer='normal', 
                activation='relu'))
model.add(tf.keras.layers.Dense(1, kernel_initializer='normal'))
# Compile model
model.compile(loss='mean_squared_error', optimizer='adam')

fit_model = nn.fit(X_train_scaled, y_train, epochs=5)
```

```python
nn.evaluate(X_test_scaled, y_test, verbose=2)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
